```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train_original.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df[df['class_name']=='No finding']['class_name'].count() #number of 'No finding' class in dataset
```

Visualization

```python
plt.figure(figsize=(26, 8))
x=sns.countplot(x="class_id", data=df)
#x.axes.set_title("Class id Distribution",fontsize=20)
x.set_xlabel("class_id",fontsize=20)
x.set_ylabel("count",fontsize=20)
x.tick_params(labelsize=15)
plt.show()
```

```python
df1=df[df['class_id']!=14] #dataframe without 'No finding' class
```

```python
df1.shape
```

```python
plt.figure(figsize=(26, 8))
p=sns.countplot(x="class_id", data=df1)
#p.axes.set_title("Distribution of class_id without class 14",fontsize=20)
p.set_xlabel("class_id",fontsize=20)
p.set_ylabel("count",fontsize=20)
p.tick_params(labelsize=15)
plt.show()
```
